```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
# Problem A.i

alphas = np.linspace(0.01, 3, 30)

lasso_weights = np.zeros(shape=(30, 9))

for i in range(30):
    lasso = Lasso(alpha=alphas[i], fit_intercept=False)
    lasso.fit(X_train, y_train)
    lasso_weights[i] = lasso.coef_

plt.figure()
plt.plot(alphas, lasso_weights)
plt.title('Model weights vs. alpha (Lasso regression)')
plt.xlabel('Alpha (regularization strength)')
plt.ylabel('Model weights')
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='right')
plt.show()
```

```python
# Problem A.ii

alphas = [i for i in range(1, 10001)]

ridge_weights = np.zeros(shape=(10000, 9))

for i in range(10000):
    ridge = Ridge(alpha=alphas[i], fit_intercept=False)
    ridge.fit(X_train, y_train)
    ridge_weights[i] = ridge.coef_

plt.figure()
plt.plot(alphas, ridge_weights)
plt.title('Model weights vs. alpha (Ridge regression)')
plt.xlabel('Alpha (regularization strength)')
plt.ylabel('Model weights')
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='right')
plt.show()
```
